```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd

# URL of the website to scrape
url = "https://naruto.fandom.com/wiki/Hinata_Hy%C5%ABga"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

response = requests.get(url, headers=headers)
response.raise_for_status()

soup = bs(response.content, 'lxml')

# Define the sections to scrape
sections = [
    'Prologue_—_Land_of_Waves',
    'Chūnin_Exams',
    'Konoha_Crush',
    'Sasuke_Recovery_Mission',
    'Bikōchū_Search_Mission',
    'Academy_Student_Lead_Mission',
    'Gosunkugi_Capture_Mission',
    'Buried_Gold_Excavation_Mission',
    'Peddlers_Escort_Mission',
    'Lady_Daimyō_Search_Mission',
    'Sunagakure_Support_Mission',
    "In_Naruto's_Footsteps:_The_Friends'_Paths",
    'Tenchi_Bridge_Reconnaissance_Mission',
    'Akatsuki_Suppression_Mission',
    "Three-Tails'_Appearance",
    'Itachi_Pursuit_Mission',
    'Fated_Battle_Between_Brothers',
    'Six-Tails_Unleashed',
    "Pain's_Assault",
    'Five_Kage_Summit',
    'Power',
    'Paradise_Life_on_a_Boat',
    'Fourth_Shinobi_World_War:_Confrontation',
    'Fourth_Shinobi_World_War:_Climax',
    "Birth_of_the_Ten-Tails'_Jinchūriki",
    'Kaguya_Ōtsutsuki_Strikes'
]

# Initialize an empty list to store descriptions
hinata_desc = []

def clean_title(title):
    return title.replace('_', ' ').replace('—', ' ').replace("'", "'").strip()

for section in sections:
    try:
        section_id = section.replace('—', ' ').replace("'", " ")
        section_tag = soup.find('span', id=section_id).parent
        next_section_tag = section_tag.find_next('h2')

        current_element = section_tag.find_next_sibling()
        while current_element and current_element != next_section_tag:
            if current_element.name == 'p':
                description = current_element.text.strip()
                if description and not description.startswith("Main article"):
                    section.replace('_', ' ').replace('—', ' ').replace("'", "'").strip()
                    hinata_desc.append([url, clean_title(section), description])
            current_element = current_element.find_next_sibling()
    except AttributeError:
        print(f"Section '{section}' not found on the page.")

# Convert the list to a DataFrame
hinata_desc_df = pd.DataFrame(hinata_desc, columns=['URL', 'Anime Section', 'Description'])

hinata_desc_df
```
