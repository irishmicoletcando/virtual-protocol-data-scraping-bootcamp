```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
from requests.adapters import HTTPAdapter
from requests.packages.urllib3.util.retry import Retry
```

# Scraping first website

```python
url = "https://thatonerule.com/rule/hinata-hyuga-quotes/"
response = requests.get(url)
response.raise_for_status()

soup = bs(response.content, 'lxml')

hinata_hyuga_quotes = []

related_hinata_quotes = soup.find_all('div', class_='content post-excerpt entry-content clearfix')

for div in related_hinata_quotes:
    p_tags = div.find_all('p')
    for p in p_tags:
        quote = p.text.strip()
        if quote:
            quote = f'"{quote}"'  # Add quotation marks around the quote
            hinata_hyuga_quotes.append([url, quote])

hinata_df = pd.DataFrame(hinata_hyuga_quotes, columns=['URL', 'Quote'])

hinata_df.to_csv('hinata_hyuga_quotes_1.csv', index=False)

hinata_df
```

# Scraping second website

```python
url = "https://shinobi-striker.fandom.com/wiki/Hinata_Hyuga/Quotes"
response = requests.get(url)
response.raise_for_status()

soup = bs(response.content, 'lxml')

hinata_hyuga_quotes_2 = []

related_hinata_quotes = soup.find_all('i')

for hinata_quote in related_hinata_quotes:
    quote = hinata_quote.text.strip()
    if quote:
        hinata_hyuga_quotes_2.append([url, quote])

hinata_df_2 = pd.DataFrame(hinata_hyuga_quotes_2, columns=['URL', 'Quote'])

hinata_df_2.to_csv('hinata_hyuga_quotes_2.csv', index=False)

hinata_df_2
```

# Scraping third website

```python
url = "https://www.quoteambition.com/hinata-hyuga-quotes/"
response = requests.get(url)
response.raise_for_status()

soup = bs(response.content, 'lxml')

hinata_hyuga_quotes_3 = []

related_hinata_quotes = soup.find_all('p')

for hinata_quote in related_hinata_quotes:
    quote = hinata_quote.text.strip()
    if quote and quote[0].isdigit():
        quote = quote.lstrip('0123456789. ')
        hinata_hyuga_quotes_3.append([url, quote])

hinata_df_3 = pd.DataFrame(hinata_hyuga_quotes_3, columns=['URL', 'Quote'])

hinata_df_3.to_csv('hinata_hyuga_quotes_3.csv', index=False)

hinata_df_3
```

# Scraping fourth website

```python
def requests_retry_session(retries=3, backoff_factor=0.3, status_forcelist=(500, 502, 504), session=None):
    session = session or requests.Session()
    retry = Retry(
        total=retries,
        read=retries,
        connect=retries,
        backoff_factor=backoff_factor,
        status_forcelist=status_forcelist,
    )
    adapter = HTTPAdapter(max_retries=retry)
    session.mount('http://', adapter)
    session.mount('https://', adapter)
    return session

url = "https://screenrant.com/naruto-hinatas-best-most-memorable-quotes/"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

response = requests_retry_session().get(url, headers=headers)
response.raise_for_status()

soup = bs(response.content, 'lxml')

hinata_hyuga_quotes_4 = []

# Find all <h3> tags
related_hinata_quotes = soup.find_all('h3')

for hinata_quote in related_hinata_quotes:
    quote = hinata_quote.text.strip()
    if quote and '"' in quote:
        hinata_hyuga_quotes_4.append([url, quote])

hinata_df_4 = pd.DataFrame(hinata_hyuga_quotes_4, columns=['URL', 'Quote'])

hinata_df_4.to_csv('hinata_hyuga_quotes_4.csv', index=False)

hinata_df_4
```

# Scraping sixth website

```python
url = "https://animemotivation.com/hinata-hyuga-quotes/"
response = requests.get(url)
response.raise_for_status()

soup = bs(response.content, 'lxml')

hinata_hyuga_quotes_6 = []

related_hinata_quotes = soup.find_all('blockquote')

for hinata_quote in related_hinata_quotes:
    par_quote = hinata_quote.find('p')
    if par_quote:
        quote = par_quote.text.strip()
        if quote:
            quote = quote.replace("– Hinata Hyuga", "").strip()
            if not (quote.startswith('"') and quote.endswith('"')):
                quote = f'"{quote}"'
            hinata_hyuga_quotes_6.append([url, quote])

hinata_df_6 = pd.DataFrame(hinata_hyuga_quotes_6, columns=['URL', 'Quote'])

hinata_df_6.to_csv('hinata_hyuga_quotes_6.csv', index=False)

hinata_df_6
```

# Scraping seventh website

```python
def requests_retry_session(retries=3, backoff_factor=0.3, status_forcelist=(500, 502, 504), session=None):
    session = session or requests.Session()
    retry = Retry(
        total=retries,
        read=retries,
        connect=retries,
        backoff_factor=backoff_factor,
        status_forcelist=status_forcelist,
    )
    adapter = HTTPAdapter(max_retries=retry)
    session.mount('http://', adapter)
    session.mount('https://', adapter)
    return session

url = "https://www.cbr.com/naruto-hinata-best-quotes/"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

response = requests_retry_session().get(url, headers=headers)
response.raise_for_status()

soup = bs(response.content, 'lxml')

hinata_hyuga_quotes_7 = []

related_hinata_quotes = soup.find_all('span')

for hinata_quote in related_hinata_quotes:
    quote = hinata_quote.text.strip()
    if quote and '"' in quote:
        hinata_hyuga_quotes_7.append([url, quote])

hinata_df_7 = pd.DataFrame(hinata_hyuga_quotes_7, columns=['URL', 'Quote'])

hinata_df_7.to_csv('hinata_hyuga_quotes_7.csv', index=False)

hinata_df_7
```

# CSV Merging

```python
# Load the two CSV files into separate DataFrames
df_quotes_1 = pd.read_csv('hinata_hyuga_quotes_1.csv')
df_quotes_2 = pd.read_csv('hinata_hyuga_quotes_2.csv')
df_quotes_3 = pd.read_csv('hinata_hyuga_quotes_3.csv')
df_quotes_4 = pd.read_csv('hinata_hyuga_quotes_4.csv')
df_quotes_5 = pd.read_csv('hinata_hyuga_quotes_5.csv')
df_quotes_6 = pd.read_csv('hinata_hyuga_quotes_6.csv')
df_quotes_7 = pd.read_csv('hinata_hyuga_quotes_7.csv')


# Concatenate the DataFrames
df_combined = pd.concat([df_quotes_1, df_quotes_2, df_quotes_3, df_quotes_4, df_quotes_5, df_quotes_6, df_quotes_7], ignore_index=True)

# Save the combined DataFrame to a new CSV file
df_combined.to_csv('hinata_combined_quotes.csv', index=False)
```
